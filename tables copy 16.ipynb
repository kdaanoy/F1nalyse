```python
# yearSchedule = fastf1.get_event_schedule(2025)
# raceSchedule = yearSchedule.get_event_by_round(22)
# session = fastf1.get_session(2020, "Nurburgring", "Race")
# session.load(laps = True, telemetry = False)
# laps = session.laps
# print(laps['LapNumber'].max())
# print(yearSchedule)
import pandas as pd

df = pd.read_csv("Session.csv")

def getRaceLaps(row):
    date = row["DateOfSession"]
    year = pd.to_datetime(date).year
    circuitID = row["CircuitID"]
    dfread = pd.read_csv("Circuit.csv")
    city = dfread[dfread["ID"] == circuitID]["City"].iloc[0]

    try:
        session = fastf1.get_session(year, str(city), "R")
        session.load(laps = True, telemetry = False)
        laps = session.laps
        print(laps['LapNumber'].max())
        return laps['LapNumber'].max()
    except Exception:
        return 0
    

df.loc[df["Type"] == "Race", "RaceLaps"] = df[df["Type"] == "Race"].apply(getRaceLaps, axis=1)
df.to_csv("Session.csv", index=False)
```

```python
import pandas as pd
from fastf1 import plotting

erg = Ergast()

dfwrite = pd.DataFrame(columns = ["ID", "TeamName", "Color", "Abbreviation"])
dfwrite.to_csv("Team.csv", index = False)

for year in range(2018, 2026):
    constructors = erg.get_constructor_info(season = year)

    for constructor in constructors['constructorName']:
        name = (constructor)
        session = fastf1.get_session(year, 1, 'R')
        color = (plotting.get_team_color(constructor, session))
        abbreviation = name.upper()[:3]
        
        row = {"ID": 0, "TeamName": name, "Color": color,  "Abbreviation": abbreviation}
        pd.DataFrame([row]).to_csv("Team.csv", mode="a", header=False, index=False)

```

```python
import pandas as pd
from fastf1 import plotting

erg = Ergast()

dfwrite = pd.DataFrame(columns = ["ID", "DriverName", "Color", "TeamID", "DriverNumber"])
dfwrite.to_csv("Drivers.csv", index = False)

for year in range(2018, 2026):
    wdc = (erg.get_driver_standings(season = year).content[0])

    for position in wdc['position']:
        driver = (wdc[wdc['position'] == position].iloc[0])
        name = driver['givenName'] + " " + driver['familyName']
        session = fastf1.get_session(year, 1, 'R')
        try:
            color = (plotting.get_driver_color(name, session))
        except Exception:
            color = "-"
        number = driver['driverNumber']
        dfread = pd.read_csv("Team.csv")

        for team in driver['constructorNames']:
            teamName = dfread[dfread["TeamName"] == team]['ID'].iloc[0]
            row = {"ID": 0, "DriverName": name, "Color": color, "TeamID": teamName, "Number": number}
            pd.DataFrame([row]).to_csv("Drivers.csv", mode="a", header=False, index=False)
```
